```python
# Histogram Graph means frequeney plot for continous variable

# Discrete and cantinous variable
# Discrete variable ==> we can be counted for example : no. of student,
# cantinous variable  ==> we cann't be counted , only measured.
# for ex. : area, price, Height, weight, speed
```

```python
# Frequency plat(Histogram Graph) for continous variable  :
```

```python
area=[10,11,14,15,18,20,21,26,27,28,29,30,31]  # unit of area in sq. feet
```

```python
# Frequence means occurance no. of occurance
# bins means interval
# these data (means area's data) dicided into 3 bins. i habe 3 interval
# bins / interval =3
# first element of area object=10 and last element of area object=31
# interval size=(last element -first element)/bins
# interval size=(31-10)/3 ==> interval size=7.0
# one interval size will be 7
# we have first interval is 10-17 [10,11,14,15]
# second interval is 17-24[18,20,21]
# third interval is 24-31 [26,27,28,29,30,31]
# how much frequency in first interval =4
# how much frequency in first interval =3
# how much frequency in first interval =6
```

```python
plt.hist(area,bins=3)
plt.show()
```

```python
plt.figure(figsize=(10,5))     # to changing the size of figure
plt.hist(area,bins=3)
plt.title('Area Distribution')
plt.xlabel('area')
plt.ylabel('frequency')
plt.xticks(rotation=60)   # to define different pattern on xlabel title   ( for decorete )
plt.yticks(rotation=60)   # to define different pattern on ylabel title  ( for decorete )
plt.show()
```

```python
data['make'].value_counts()
```

```python
plt.figure(figsize=(10,5))     # to changing the size of figure
data['make'].value_counts().plot(kind='bar')
plt.title('company dedail')
plt.xlabel('company name')
plt.ylabel('number of care')
# plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(10,5))     # to changing the size of figure
data['make'].value_counts().plot(kind='barh')   # here h means horisontenl
plt.title('company dedail')
plt.xlabel('company name')
plt.ylabel('number of care')
plt.show()
```
